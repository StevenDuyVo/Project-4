```python
# distribution of the classes in training dataset
targets = data.HeartDiseaseorAttack.value_counts()
colors = ['pink','red']

# visualization for the bar chart
figure, axes = plt.subplots(1,2, figsize=(15,5))
axes[0].barh(y=targets.index, width=targets.values, color=colors)
axes[0].set_xlabel('Frequency')
axes[0].set_ylabel('Heart Diseases')
axes[0].set_yticks([0,1], ['Healthy','Heart Disease'])


for index, values in enumerate(targets):
    axes[0].text(values+1000, index, str(values), va='center')

# visualization for the pie chart    
axes[1].pie(targets.values, labels=['Healthy','Heart Disease'], autopct='%.2f%%', explode=([0.05]*len(targets.index)), colors=colors)

#Title of chart
figure.suptitle('Distribution of Samples: Healthy vs Heart Disease', fontsize=15)

#Show the Chart
plt.show()
```

```python
from scipy.stats import chi2_contingency

chisquare_scores = []
for column in data.columns:
    contingency_table = pd.crosstab(data[column], data['HeartDiseaseorAttack'])
    chi2, p, dof, expected = chi2_contingency(contingency_table,)
    chisquare_scores.append({'Attributes': column, 'Chi-square score': chi2})

chisquare_df = pd.DataFrame(chisquare_scores)
chisquare_df
```

```python
sorted_df = chisquare_df.sort_values(by='Chi-square score', ascending=False)
sorted_df=sorted_df.drop(0)
plt.bar(sorted_df['Attributes'], sorted_df['Chi-square score'], color='Red')
plt.xlabel('Attributes')
plt.ylabel('Chi-square score')
plt.title('Chi-square Scores for Each Attribute')
plt.xticks(rotation=90)
```
